```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('https://github.com/ShivaniBandapalli/Movie-Recommendation/raw/main/dataset.csv')
```

```python
movies.head()
```

```python
movies.columns
```

```python
movies.info()
```

```python
movies['tags'] = movies['genres'] + movies['overview']
```

```python
movies.head()
```

```python
new_df = movies[['id','original_title','genres','overview','tags']]
```

```python
new_df = new_df.drop(columns=['genres','overview'])
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vec = cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
```

```python
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vec)
```

```python
new_df[new_df['original_title']=='Jurassic World']
```

```python
dist = sorted(list(enumerate(sim[0])),reverse=True,key=lambda vec:vec[1])
```

```python
dist
```

```python
for i in dist[0:5]:
   print(new_df.iloc[i[0]].original_title)
```

```python
def recommend(movies):
  index = new_df[new_df['original_title']==movies].index[0]
  distance = sorted(list(enumerate(sim[index])),reverse=True,key=lambda vec:vec[1])
  for i in distance[0:5]:
    print(new_df.iloc[i[0]].original_title)
```

```python
recommend("Iron Man")
```
